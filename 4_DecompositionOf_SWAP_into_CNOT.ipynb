Find the decomposition of SWAP gate into CNOT gate. Write a program that shows that both the circuits perform same operation by using the class 'state vector' from qiskit.quantum_info

```python
import numpy as np

# Importing standard Qiskit libraries
from qiskit import QuantumCircuit, transpile, Aer, IBMQ, QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit.tools.jupyter import *
from qiskit.visualization import *
from ibm_quantum_widgets import *
from qiskit.providers.aer import QasmSimulator
import qiskit

# Loading your IBM Quantum account(s)
provider = IBMQ.load_account()
```

```python
#Decomposition of a swapgate is 3 cnot gates one from q0-q1 one for q1-q0 and one from q0-q1
#we can visualise this using the statevector method in the quantum_info module

import qiskit.quantum_info as qi

q=qiskit.QuantumRegister(2,'q')

circuit_a=QuantumCircuit(q)
circuit_a.x(q[1])
circuit_a.h(q[0])
circuit_a.z(q[1])
circuit_a.z(q[0])
circuit_a.cx(q[0],q[1])

print('Original Circuit (4th Bell State):')
circuit_a.draw()
```

```python
circuit_b=QuantumCircuit(q)
circuit_b.x(q[0])
circuit_b.h(q[1])
circuit_b.z(q[0])
circuit_b.z(q[1])
circuit_b.cx(q[1],q[0])

print('Swap Circuit (swapped 4th Bell State):')
circuit_b.draw()
```

```python
qi.Statevector(circuit_a).draw('city')
```

```python
qi.Statevector(circuit_b).draw('city')
```

```python
circuit_c=QuantumCircuit(q)
circuit_c.x(q[1])
circuit_c.h(q[0])
circuit_c.z(q[1])
circuit_c.z(q[0])
circuit_c.cx(q[0],q[1])

circuit_c.cx(q[0],q[1])
circuit_c.cx(q[1],q[0])
circuit_c.cx(q[0],q[1])

print('Circuit + SWAP = Swapped Circuit')
circuit_c.draw()
```

```python
qi.Statevector(circuit_c).draw('city')
```
